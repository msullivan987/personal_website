```python
import json
from pprint import pprint
import pandas as pd
from datetime import datetime
```

```python
# Load JSON

with open("2020_APRIL.json") as jsonfile:
    location_json = json.load(jsonfile)
```

```python
# print(pprint(location_json["timelineObjects"]))
```

```python
timelineObjects = location_json["timelineObjects"]
```

```python
# Activity Data 
activities = []
activity_start_time = []
activity_end_time = []

# Place visit data
locations = []
visit_start_time = []
visit_end_time = []

for entry in timelineObjects:   
    try:
        activities.append(entry["activitySegment"]["activityType"])
        activity_start_time.append(entry["activitySegment"]["duration"]["startTimestampMs"])
        activity_end_time.append(entry["activitySegment"]["duration"]["endTimestampMs"])
    except KeyError:
        locations.append(entry["placeVisit"]["location"]["name"])
        visit_start_time.append(entry["placeVisit"]["duration"]["startTimestampMs"])
        visit_end_time.append(entry["placeVisit"]["duration"]["endTimestampMs"])
        
activity_start_time = pd.to_numeric(activity_start_time)
activity_end_time = pd.to_numeric(activity_end_time)
visit_start_time = pd.to_numeric(visit_start_time)
visit_end_time = pd.to_numeric(visit_end_time)
```

```python
locations_df = pd.DataFrame({
    "location_name":locations,
    "visit_start_time":visit_start_time,
    "visit_end_time":visit_end_time
})

activities_df = pd.DataFrame({
    "activity":activities,
    "activity_start_time":activity_start_time,
    "activity_end_time":activity_end_time
})

```

```python
time_elapsed_location = (locations_df["visit_end_time"] - locations_df["visit_start_time"])/1000/60

locations_df["Time at location (minutes)"] = time_elapsed_location

locations_df["Datetime"] = locations_df["visit_start_time"].astype('float64').apply(lambda x: datetime.utcfromtimestamp(x/1000))

locations_df.head()
```

```python
grouped_locations = locations_df.groupby("location_name").sum().drop(columns=["visit_start_time", "visit_end_time"])
grouped_locations
```

```python
time_elapsed_activity = (activities_df["activity_end_time"] - activities_df["activity_start_time"])/1000/60

activities_df["Length of Activity (minutes)"] = time_elapsed_activity

activities_df.head(20)
```
